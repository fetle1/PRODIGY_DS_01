```python
import plotly.express as px
```

```python
histogram = px.histogram(
    df,
    x='age',  
    title='Distribution of Age',
    labels={'age': 'Age (years)'},
    color_discrete_sequence=['#636EFA'],  
)

histogram.show()
```

```python
histogram = px.histogram(
    df, 
    x='age',  
    nbins=30,  
    title='Distribution of Age',
    labels={'age': 'Age (years)'},
    color_discrete_sequence=['#40E0D0'],  
)

histogram.update_layout(
    showlegend=False,  
    coloraxis_showscale=False 
)

histogram.show()
```

```python
 age_counts = df['age'].value_counts().sort_index().reset_index()
age_counts.columns = ['Age', 'Count']  


bar_chart = px.bar(
    age_counts,
    x='Age',
    y='Count',
    title='Distribution of Age (All Categories)',
    labels={'Age': 'Age (years)', 'Count': 'Count'},
    color='Count',  
    color_continuous_scale='Viridis',  
    color_continuous_midpoint=age_counts['Count'].mean()  
)

bar_chart.update_layout(
    showlegend=True,
    coloraxis_colorbar=dict(title="Count")  
)

bar_chart.show()
```

```python
bmi_labels = ['Underweight', 'Normal', 'Overweight', 'Obese', 'Morbid Obesity']
df['BMI_category'] = pd.cut(
    df['BMI'],
    bins=[0, 18.5, 24.9, 29.99, 34.9, float('inf')],
    labels=bmi_labels
)

# Get the count of each BMI category
bmi_counts = df['BMI_category'].value_counts().reset_index()
bmi_counts.columns = ['BMI Category', 'Count']  # Rename columns for clarity

# Bar Chart for the 'BMI_category' variable
bar_chart = px.bar(
    bmi_counts,
    x='BMI Category',
    y='Count',
    title='Distribution of BMI Categories',
    labels={'BMI Category': 'BMI Category', 'Count': 'Count'},
    color='BMI Category',
    color_discrete_sequence=px.colors.qualitative.Set3  # Custom color scheme
)

bar_chart.update_layout(showlegend=False)
bar_chart.show()
bar_chart.write_html("interactive_BMI_bar grapgh.html")
```

```python
bmi_labels = ['Underweight', 'Normal', 'Overweight', 'Obese', 'Morbid Obesity']
df['BMI_category'] = pd.cut(
    df['BMI'],
    bins=[0, 18.5, 24.9, 29.99, 34.9, float('inf')],
    labels=bmi_labels
)

# Bar Chart for Categorical Variable: 'BMI_category'
bar_chart = px.bar(
    df,
    x='BMI_category',
    title='Distribution of BMI Categories',
    labels={'BMI_category': 'BMI Category'},
    color='BMI_category'
)
bar_chart.update_layout(showlegend=False)
bar_chart.show()

# Histogram for Continuous Variable: 'age'
histogram = px.histogram(
    df,
    x='age',
    title='Distribution of Age',
    nbins=10,
    color_discrete_sequence=['#AB63FA']  # Custom color
)
histogram.update_layout(
    xaxis_title='Age (years)',
    yaxis_title='Count'
)
histogram.show()
```
