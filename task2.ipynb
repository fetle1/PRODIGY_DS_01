```python
import plotly.express as px
import plotly.graph_objects as go

# Get the count of each BMI category
bmi_counts = df['BMI_category'].value_counts().reset_index()
bmi_counts.columns = ['BMI Category', 'Count']  # Rename columns for clarity

# Create the initial bar chart
fig = go.Figure()

# Add the initial bar chart with empty data
fig.add_trace(go.Bar(
    x=bmi_counts['BMI Category'],
    y=[0] * len(bmi_counts),  # Start with no bars visible
    name='BMI Categories',
    marker=dict(color=px.colors.qualitative.Set3),
))

# Create frames to animate the bars
frames = [
    go.Frame(
        data=[
            go.Bar(
                x=bmi_counts['BMI Category'],
                y=bmi_counts['Count'].iloc[:k],  # Show up to the k-th category
                marker=dict(color=px.colors.qualitative.Set3),
            )
        ],
        name=f'frame{k}'  # Name of the frame for each animation step
    )
    for k in range(1, len(bmi_counts) + 1)
]

# Add the frames to the figure
fig.frames = frames

# Add the layout for the figure
fig.update_layout(
    title="Distribution of BMI Categories (Animated)",
    xaxis_title="BMI Category",
    yaxis_title="Count",
    showlegend=False,
    updatemenus=[{
        "buttons": [
            {
                "args": [None, {"frame": {"duration": 500, "redraw": True}, "fromcurrent": True}],
                "label": "Play",
                "method": "animate",
            },
            {
                "args": [[None], {"frame": {"duration": 0, "redraw": True}, "mode": "immediate", "transition": {"duration": 0}}],
                "label": "Pause",
                "method": "animate",
            }
        ],
        "direction": "left",
        "pad": {"r": 10, "t": 87},
        "showactive": False,
        "type": "buttons",
        "x": 0.1,
        "xanchor": "right",
        "y": 0,
        "yanchor": "top"
    }],
    sliders=[{
        "steps": [
            {
                "args": [
                    [f"frame{k}"], 
                    {"frame": {"duration": 500, "redraw": True}, "mode": "immediate", "transition": {"duration": 0}},
                ],
                "label": str(k),
                "method": "animate",
            }
            for k in range(1, len(bmi_counts) + 1)
        ]
    }],
)

# Show the animation
fig.show()

# Save the animation as HTML (which retains interactivity)
fig.write_html("animated_BMI_distribution.html")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

variables = ['cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']
for var in variables:
    plt.figure(figsize=(8, 6))
    sns.regplot(x=var, y='age', data=df, scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
    plt.title(f'Scatter Plot of {var} vs Age with Regression Line')
    plt.xlabel(var)
    plt.ylabel('Age')
    plt.grid(True)
    plt.show()
```
